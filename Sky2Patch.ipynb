```python
from ccgpack import sky2patch
```

```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
```

```python
 cl=np.load('/content/drive/MyDrive/Colab Notebooks/zahra project/cl_planck_lensed.npy')
```

```python
ll = cl[ :,0]
cl = cl[ :,1]
```

```python
sky=hp.sphtfunc.synfast(cl, nside=2048)
```

```python
hp.mollview(sky, cmap='jet')
```

```python
patch = sky2patch(sky , npatch=12)
print(patch.shape)
```

# The first 24 patches:






```python
fig, ax = plt.subplots (4, 6, figsize=[12,12])
plt.subplots_adjust(bottom=0.3, top=0.8, wspace=0, hspace=0)

for i in range(4):
    for j in range(6):
        ax[i,j].imshow(patch[(i)*(j+1)], cmap='jet')
        #ax[i,j].set_title (str((i+1)*(j+1)))
        ax[i,j].set_xticks([])
        ax[i,j].set_yticks([])
```

# Random sample 

```python
seeds=range(1728)
```

```python

fig, ax = plt.subplots (4, 6, figsize=[12,12])
plt.subplots_adjust(bottom=0.3, top=0.9, wspace=0, hspace=0)

for i in range(4):
    for j in range(6):
        rnd=np.random.choice(seeds , size=1)
        ax[i,j].imshow(patch[int(rnd)], cmap='jet')
        ax[i,j].set_title (str(int(rnd)))
        ax[i,j].set_xticks([])
        ax[i,j].set_yticks([])
```
